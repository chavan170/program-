```python
#Q1.Write a python program to calculate student grade using elif statement and display it with student details.
name = input("Enter student's name: ")
PRN = input("Enter student's PRN: ")
branch = input("Enter student's branch: ")
sub1 = int(input("Enter marks of 1st Subject:"))
sub2 = int(input("Enter marks of 2nd Subject:"))
sub3 = int(input("Enter marks of 3rd Subject:"))

total= sub1 + sub2 + sub3
percentage = float((total/300) * 100)

if percentage >= 90:
    grade = 'A'
elif percentage >= 80:
    grade = 'B'
elif percentage >= 70:
    grade = 'C'
elif percentage >= 60:
    grade = 'D'
else:
    grade = 'F'

print("\n Student Details:")
print("Name:", name)
print("PRN:", PRN)
print("Branch:", branch)
print("Total Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)


```

```python
#Q3. Write a python program to print alphabet L and Z patterns using nested for loop.

# L pattern
print("Alphabet L Pattern:")
for i in range(1, 6):
    if i != 5:
        print("*")
    else:
        print("*" * 5)
# Z pattern
print("\nAlphabet Z Pattern:")
for i in range(5):
    if i == 0 or i == 4:
        print("*" * 5)
    else:
        print(" " * (4 - i) + "*")
```

```python
#Q5.Write a python program to calculate and display electricity bill using elif statement.

name=input("Enter Consumer name: ")
id=int(input("Enter Consumer Number: "))
units=int(input("Enter units: "))

if units<100:
    Amount=units*1.5
    charges=25.00

elif units<=200:
    Amount=(100*1.5)+(units-100)*2.5
    charges=50.00

elif units<=300:
    Amount=(100*1.5)+(200-100)*2.5+(units-200)*4
    charges=75.00

elif units<=350:
    Amount=(100*1.5)+(200-100)*2.5+(300-200)*4+(units-300)*5
    charges=100.00
else:
    Amount=0
    charges=1500.00
total=Amount+charges

print("Consumer Name:  ", name)
print("Consumer Number: ", id)
print("Electricity bill for month: ",total)
print("Pay Electricity Bill before due date.\n")
print("*****  Have a Nice Day !!!  *****")
```
